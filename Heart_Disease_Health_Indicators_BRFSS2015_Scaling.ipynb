```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')
df
```

```python
df.columns
```

```python
df1= df
df1.dtypes
```

```python
df_basic = pd.DataFrame(index = df1.columns, columns=['Min', 'Max', 'Sum', 'Mean', 'Mode'])
df_basic['Min'] = df1.min()
df_basic['Max'] = df1.max()
df_basic['Sum'] = df1.sum()
df_basic['Mean'] = df1.mean()
df_basic['Mode'] = df1.mode().T
df_basic
```

```python
#Maximum absolute scaling
df_max_scaled = df1

for column in df_max_scaled.columns:
    df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()
df_max_scaled
```

```python
#min-max scaling
df_min_max_scaled = df1

for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

df_min_max_scaled
```

```python
#z-score
z_scaled = df1
for column in z_scaled.columns:
    z_scaled[column] = (z_scaled[column] - z_scaled[column].mean()) / z_scaled[column].std()

z_scaled
```
